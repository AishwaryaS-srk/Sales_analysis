# Exploratory Data Analysis

#### Gender

```python
df.columns
```

```python
sns.countplot(x='Gender',data = df)
```

```python
ax = sns.countplot(x='Gender',data = df)
for bar in ax.containers:
    ax.bar_label(bar)
```

The above graph shows that female count of purchasing the product is more than males.

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x='Gender',y='Amount',data = sales_gen)
```

The purchasing power(cost) of females are greater than males

#### Age

```python
df.columns
```

```python
ax = sns.countplot(data = df,x='Age Group',hue='Gender')
for bar in ax.containers:
    ax.bar_label(bar)
```

Above graph shows that people(both female and male) between the age group 26-35 has more puchasing power(count)

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
ax=sns.barplot(x='Age Group',y='Amount',data = sales_age)
for bar in ax.containers:
    ax.bar_label(bar)
```

Above graph shows that people between the age group 26-35 has more puchasing power(cost)

# State

```python
df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(3)
```

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending = False).head(10)
sns.set(rc={'figure.figsize':(17,5)})  #to set the space and height around x and y axis
ax=sns.barplot(data = sales_state,x='State',y='Orders')
for bar in ax.containers:
    ax.bar_label(bar)
```

The above bar graph shows that number of purchases are more in uttar pradesh compared to other states

```python
sales_state = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_state,x='State',y='Amount')
```

The above bar graph shows that the purchase power(cost) is more in uttar pradesh

```python
ax= sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(1,5)})
for bar in ax.containers:
    ax.bar_label(bar)
```

The above graph shows that the people with marital_status as single has made a lot of purchase 

```python
sales_state = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state,x="Marital_Status",y='Amount',hue='Gender')
```

The above bar graph shows that both male and female with marital status as single spent more on their purchase

#### Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df,x='Occupation')
for bar in ax.containers:
    ax.bar_label(bar)
```

The graph above shows that the people working in the IT sector has made more no.of purchases

```python
sales_state = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state,x='Occupation',y='Amount')
```

The above figure shows that people working in IT sector has spent a lot on their purchase

# Product Category

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df,x='Product_Category')
for bar in ax.containers:
    ax.bar_label(bar)
```

The above graph shows that clothing and apparel products got sold more

```python
sales_state = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,15)})
sns.barplot(data = sales_state,x='Product_Category',y='Amount')
```

The graph above shows that the amount spent on food is more 

```python
sales_state = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
fc=sns.barplot(data = sales_state,x='Product_ID',y='Orders')
for bar in fc.containers:
    fc.bar_label(bar)
```

The graph mentioned above displays the product_ID "P00265242" got many orders
